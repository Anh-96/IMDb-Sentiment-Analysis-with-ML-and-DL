```python
def pick_emojis(text: str):
    """
    Pick emojis from a piece of text
    :param text: string
    :return: emojis
    """
    text = str(text)
    emojis = []

    for char in text:
        if char in emoji.EMOJI_DATA:
            emojis.append(char)

    return "".join(emojis)
```

```python
df['only_emojis'] = df['review'].apply(pick_emojis)
```

```python
df['only_emojis'].value_counts()
```
